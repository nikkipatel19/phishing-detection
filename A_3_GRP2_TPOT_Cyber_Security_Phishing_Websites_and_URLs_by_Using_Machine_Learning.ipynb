```python
# function for finding the correlation between the features...
def corr_heatmap(data, idx_s, idx_e):
    y = data['CLASS_LABEL']
    temp = data.iloc[:, idx_s:idx_e]
    if 'id' in temp.columns:
        del temp['id']
    temp['CLASS_LABEL'] = y
    sns.heatmap(temp.corr(), annot=True, fmt='.2f')

    plt.show()
```

```python
# visualizing the correlation graph...
corr_heatmap(df, 0, 10)
```

```python
# splitting the dependent and independent variable...
from sklearn.feature_selection import mutual_info_classif
X = df.drop(['id', 'CLASS_LABEL'], axis=1)
y = df['CLASS_LABEL']
```

```python
discrete_features = X.dtypes == int
mi_scores = mutual_info_classif(X, y, discrete_features=discrete_features)
mi_scores = pd.Series(mi_scores, name='MI Scores', index=X.columns)
mi_scores = mi_scores.sort_values(ascending=False)
mi_scores
```

```python
# function of calculating the mutual information scores and visualizing it...
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("MI Scores")

plt.figure(dpi=100, figsize=(12,12))
plot_mi_scores(mi_scores)
```

```python
#phishing and non phishing distribution
Phishing = df[df["CLASS_LABEL"] == 1]
valid = df[df["CLASS_LABEL"] == 0]
```

```python
# Differentiating target variable
X = df.iloc[:,:-2].values
y = df.iloc[:, -2].values
```

```python
# Encoding the data
label_encoder_x= LabelEncoder()
X[:, 3] = label_encoder_x.fit_transform(X[:,3])
X[:, 5] = label_encoder_x.fit_transform(X[:,5])
```

```python
one_hot_enc = OneHotEncoder()

one_hot_enc.fit(X[:,1].reshape(-1,1))
one_hot_enc.categories_
X_2 = X[:,0].shape
X = np.concatenate((X[:,0].reshape(-1,1),one_hot_enc.transform(X[:,1].reshape(-1,1)).toarray(), X[:,2:]), axis=1)
```

```python
standard_sc = StandardScaler()
X = standard_sc.fit_transform(X)
```

```python
# Splitting the dataset into training and testing dataset (80-20)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
```

```python
X_50 = X_train[:50].copy()
y_50 = y_train[:50].copy()
```

```python
#tpot = TPOTClassifier(generations=3, population_size=5, verbosity=1, scoring='accuracy',cv = 3, random_state=None, n_jobs=1, max_time_mins=None)
```

```python
#tpot = TPOTClassifier(generations=10, population_size=20, verbosity=5, scoring='accuracy',cv = 10, random_state=None, n_jobs=1, max_time_mins=None)
```

```python
tpot = TPOTClassifier(generations=5, population_size=10, verbosity=3, scoring='accuracy',cv = 5, random_state=None, n_jobs=1, max_time_mins=None)
```

```python
tpot.fit(X_50, y_50)
predictions = tpot.predict(X_test)
score = tpot.score(X_test, y_test)

print(tpot.fitted_pipeline_)
tpot.export('tpot_best_pipeline.py')
```

```python
tpot.fitted_pipeline_
```

```python
from sklearn.metrics import confusion_matrix, classification_report


predictions = tpot.predict(X_test)
cm = confusion_matrix(y_test, predictions)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=tpot.classes_, yticklabels=tpot.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

report = classification_report(y_test, predictions, zero_division=1)  # Use zero_division parameter to handle warnings
print("Classification Report:\n", report)

final_model = tpot.fitted_pipeline_.steps[-1][1]
print("Final Fitted Pipeline:", final_model)

accuracy = tpot.score(X_test, y_test)
print("Accuracy on test set:", accuracy)
```
